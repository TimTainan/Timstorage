```python
#1.載入tkinter
import tkinter as tk
import requests
import tkinter.messagebox
from datetime import datetime

def _radar():
    radar_url = 'https://opendata.cwa.gov.tw/fileapi/v1/opendataapi/O-A0058-003?Authorization=rdec-key-123-45678-011121314&format=JSON'
    radar = requests.get(radar_url)
    radar_json = radar.json()
    radar_img = radar_json['cwaopendata']['dataset']['resource']['ProductURL']
    radat_time = radar_json['cwaopendata']['dataset']['DateTime']   # 取得時間
    #print(radar_img)
    
    keY=enteR.get()
    
    d1 = datetime.now() 
    date_string1 = d1.strftime("%Y/%m/%d %H:%M(%p)")

    url = 'https://notify-api.line.me/api/notify'
    token = keY
    headers = {
        'Authorization': 'Bearer ' + token
    }
    data = {
        'message':date_string1+" "+'從雷達回波看看會不會下雨～',
        'imageThumbnail':radar_img + '?' + radat_time,    # 加上時間參數
        'imageFullsize':radar_img + '?' + radat_time      # 加上時間參數
    }
    data = requests.post(url, headers=headers, data=data)
    
def _weatherNow():
    url1 = 'https://opendata.cwa.gov.tw/fileapi/v1/opendataapi/O-A0001-001?Authorization=CWA-89F101CF-82F8-4ECC-89A9-980C8F9E75B1&downloadType=WEB&format=JSON'
    data = requests.get(url1)
    data_json = data.json()
    location = data_json['cwaopendata']['dataset']['Station']
    weather = {}   # 新增一個 weather 字典
    for i in location:
        name = i['StationName'] # 測站地點
        city = i['GeoInfo']['TownName']  # 城市
        area = i['GeoInfo']['CountyName']  # 行政區
        temp = i['WeatherElement']['AirTemperature']  # 氣溫
        humd = abs(float(i['WeatherElement']['RelativeHumidity'])) # 相對濕度
        r24 = abs(float(i['WeatherElement']['Now']['Precipitation']))  # 累積雨量
        weaTher=i['WeatherElement']['Weather']
        msg = f'{weaTher}天，氣溫{temp} 度，相對濕度 {humd}%，累積雨量 {r24}mm'  # 組合成天氣描述
        try:
            weather[city][name]=msg   # 記錄地區和描述
        except:
            weather[city] = {}        # 如果每個縣市不是字典，建立第二層字典
            weather[city][name]=msg   # 記錄地區和描述
            
    d1 = datetime.now() 
    date_string1 = d1.strftime("%Y/%m/%d %H:%M(%p)")

    aa=weather['安平區']['安平']
    
    keY=enteR.get()
    
    url = 'https://notify-api.line.me/api/notify'
    token = keY
    headers = {
        'Authorization': 'Bearer ' + token
    }
    data = {
        'message':date_string1+" "+"現在安平天氣"+aa,
    }
    data = requests.post(url, headers=headers, data=data)

    
def _aqi():
    url = 'https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=ImportDate%20desc&format=JSON'
    data = requests.get(url)             # 使用 get 方法透過空氣品質指標 API 取得內容
    data_json = data.json()              # 將取得的檔案轉換為 JSON 格式
    location = data_json['records']
    aQI = {}   # 新增一個 weather 字典
    for i in location:
        name = i['sitename'] # 測站地點
        area = i['county']  # 行政區
        aqi = i['aqi']  # aqi
        status = i['status'] # 空氣品質
        msg = f' AQI :{aqi}，空氣品質為 {status}'  
        try:
            aQI[area][name]=msg   # 記錄地區和描述
        except:
            aQI[area] = {}        # 如果每個縣市不是字典，建立第二層字典
            aQI[area][name]=msg   # 記錄地區和描述
    
    d1 = datetime.now() 
    date_string1 = d1.strftime("%Y/%m/%d %H:%M(%p)")
    
    keY=enteR.get()

    #print("現在臺南",aQI['臺南市']['臺南'])
    bb=aQI['臺南市']['臺南']

    radar_url = 'https://opendata.cwa.gov.tw/fileapi/v1/opendataapi/O-A0058-003?Authorization=rdec-key-123-45678-011121314&format=JSON'
    radar = requests.get(radar_url)
    radar_json = radar.json()
    radar_img = radar_json['cwaopendata']['dataset']['resource']['ProductURL']
    radat_time = radar_json['cwaopendata']['dataset']['DateTime']   # 取得時間

    url = 'https://notify-api.line.me/api/notify'
    token = keY
    headers = {
        'Authorization': 'Bearer ' + token
    }
    data = {
        'message':date_string1+" "+"現在臺南"+bb,
    }
    data = requests.post(url, headers=headers, data=data)

def _hit4():
    qQ=tk.messagebox.askokcancel('提示','確定要結束程式嗎?')
    if qQ:
        wiN.destroy()

#2.使用tkinter建立一個叫[wiN]的視窗
wiN=tk.Tk()

#3.設定視窗[標題]
wiN.title("Weather for Line")

#4.設定視窗[大小]
wiN.geometry("800x600")


lbL1=tk.Label(wiN,text="請輸入Line token",fg="black",bg="white",font=("Arial",16))
lbL1.pack()

enteR=tk.Entry(wiN,font=("Arial",20), width=46)
enteR.pack()

btN1=tk.Button(wiN,text="雷達迴波",font=("Arial",14),width=12,height=3,command=_radar)
btN2=tk.Button(wiN,text="安平天氣",font=("Arial",14),width=12,height=3,command=_weatherNow)
btN3=tk.Button(wiN,text="空氣品質",font=("Arial",14),width=12,height=3,command=_aqi)
btN4=tk.Button(wiN,text="結束程式",bg="gray",font=("Arial",14),width=12,height=3,command=_hit4)

btN1.pack()
btN2.pack()
btN3.pack()
btN4.pack()

#5.顯示視窗
wiN.mainloop()
```
